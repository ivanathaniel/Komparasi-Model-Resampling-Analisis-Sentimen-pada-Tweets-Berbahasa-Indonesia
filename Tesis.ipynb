## Preprocessing data text

```python
import re
import string
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factorystemmer = StemmerFactory()
stemmer = factorystemmer.create_stemmer()

factorystop = StopWordRemoverFactory()
stopwords = factorystop.create_stop_word_remover()

def clean_text(text):
    text = text.lower() #ubah text menjadi lower case
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text) #hapus tanda baca
    text = re.sub('[‘’“”…]', '', text) #hapus tanda baca lain
    text = re.sub('[0-9]+', '', text) #hapus angka
    text = re.sub('(?:a{0,2}h{1,2}a{0,2}){2,}h?', '', text) #hapus haha-ahah pattern
    text = re.sub('(?:e{0,2}h{1,2}e{0,2}){2,}h?', '', text) #hapus hehe-eheh pattern
    text = re.sub('(?:i{0,2}h{1,2}i{0,2}){2,}h?', '', text) #hapus hihi-ihih pattern
    text = re.sub(r'\b[wk]*(?:wk|kw)[wk]*\b', '', text) #hapus wkwk-kwkw pattern
    text = re.sub(r'\s—\s', '', text) #hapus spasi berlebih
    text = re.sub(r'http\S+', '', text) #hapus link
    text = re.sub('maranatha', '', text) #hapus keyword input maranatha
    text = re.sub('marnath', '', text) #hapus keyword input marnath
    text = re.sub('marnat', '', text) #hapus keyword input marnat
    #text = stemmer.stem(text) #stemming
    text = stopwords.remove(text) #hapus stopwords
    return text
```

```python
text_cleaner = []

for text in join_df.Text:
    text_cleaner.append(clean_text(text))

new_text = pd.DataFrame({'New_Text' : text_cleaner})
clean_df = pd.concat([join_df, new_text], axis = 1)
clean_df = clean_df.dropna()
clean_df
```

## Buat classifier dengan naive bayes dengan fitur tf-idf dan count vectorizer

```python
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

tvec1 = TfidfVectorizer(ngram_range=(1, 1)) #tfidf unigram
tvec2 = TfidfVectorizer(ngram_range=(1, 2)) #tfidf bigram
tvec3 = TfidfVectorizer(ngram_range=(1, 3)) #tfidf trigram
cvec1 = CountVectorizer(ngram_range=(1, 1)) #count vectorizer unigram
cvec2 = CountVectorizer(ngram_range=(1, 2)) #count vectorizer bigram
cvec3 = CountVectorizer(ngram_range=(1, 3)) #count vectorizer trigram
mnb = MultinomialNB()
cnb = ComplementNB()
```

```python
from sklearn.pipeline import Pipeline

#pipeline multinomialNB dengan tfidf unigram
mnb_tvec1_pipeline = Pipeline([
    ('vectorizer', tvec1),
    ('classifier', mnb)
])

#pipeline multinomialNB dengan tfidf bigram
mnb_tvec2_pipeline = Pipeline([
    ('vectorizer', tvec2),
    ('classifier', mnb)
])

#pipeline multinomialNB dengan tfidf trigram
mnb_tvec3_pipeline = Pipeline([
    ('vectorizer', tvec3),
    ('classifier', mnb)
])

#pipeline multinomialNB dengan count vectorizer unigram
mnb_cvec1_pipeline = Pipeline([
    ('vectorizer', cvec1),
    ('classifier', mnb)
])

#pipeline multinomialNB dengan count vectorizer bigram
mnb_cvec2_pipeline = Pipeline([
    ('vectorizer', cvec2),
    ('classifier', mnb)
])

#pipeline multinomialNB dengan count vectorizer trigram
mnb_cvec3_pipeline = Pipeline([
    ('vectorizer', cvec3),
    ('classifier', mnb)
])
```

```python
#pipeline complementNB dengan tfidf unigram
cnb_tvec1_pipeline = Pipeline([
    ('vectorizer', tvec1),
    ('classifier', cnb)
])

#pipeline complementNB dengan tfidf bigram
cnb_tvec2_pipeline = Pipeline([
    ('vectorizer', tvec2),
    ('classifier', cnb)
])

#pipeline complementNB dengan tfidf trigram
cnb_tvec3_pipeline = Pipeline([
    ('vectorizer', tvec3),
    ('classifier', cnb)
])

#pipeline complementNB dengan count vectorizer unigram
cnb_cvec1_pipeline = Pipeline([
    ('vectorizer', cvec1),
    ('classifier', cnb)
])

#pipeline complementNB dengan count vectorizer bigram
cnb_cvec2_pipeline = Pipeline([
    ('vectorizer', cvec2),
    ('classifier', cnb)
])

#pipeline complementNB dengan count vectorizer trigram
cnb_cvec3_pipeline = Pipeline([
    ('vectorizer', cvec3),
    ('classifier', cnb)
])
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix

#MultinomialNB classifier
def mnb_cv(splits, X, Y, pipeline, average_method):
    
    kfoldmnb = StratifiedKFold(n_splits=splits, shuffle=True, random_state=5)
    accuracymnb = []
    precisionmnb = []
    recallmnb = []
    f1mnb = []
    for train, test in kfoldmnb.split(X, Y):
        mnb_fit = pipeline.fit(X[train], Y[train])
        predictionmnb = mnb_fit.predict(X[test])
        scoresmnb = mnb_fit.score(X[test],Y[test])
        class_names = ['Negative','Neutral','Positive']
        cm = confusion_matrix((Y[test]), predictionmnb)
        mnbdf_cm = pd.DataFrame(cm, index=class_names, columns=class_names)
        
    
        accuracymnb.append(scoresmnb * 100)
        print('Accuracy: ', scoresmnb * 100)
        precisionmnb.append(precision_score(Y[test], predictionmnb, average=average_method)*100)
        print(' '*10, class_names)
        print('Precision:',precision_score(Y[test], predictionmnb, average=None))
        recallmnb.append(recall_score(Y[test], predictionmnb, average=average_method)*100)
        print('Recall:   ',recall_score(Y[test], predictionmnb, average=None))
        f1mnb.append(f1_score(Y[test], predictionmnb, average=average_method)*100)
        print('F1 score: ',f1_score(Y[test], predictionmnb, average=None))
        print('-'*50)

        mnbaccumean = np.mean(accuracymnb)
        mnbprecmean = np.mean(precisionmnb)
        mnbrecamean = np.mean(recallmnb)
        mnbf1scmean = np.mean(f1mnb)
            
    print('Accuracy: %.2f%% (+/- %.2f%%)' % (mnbaccumean, np.std(accuracymnb)))
    print('Precision: %.2f%% (+/- %.2f%%)' % (mnbprecmean, np.std(precisionmnb)))
    print('Recall: %.2f%% (+/- %.2f%%)' % (mnbrecamean, np.std(recallmnb)))
    print('F1 score: %.2f%% (+/- %.2f%%)' % (mnbf1scmean, np.std(f1mnb)))
    print(' ')
    print('Confusion matrix\n', mnbdf_cm)
    return kfoldmnb, accuracymnb, precisionmnb, recallmnb, f1mnb
        
#ComplementNB Classifier
def cnb_cv(splits, X, Y, pipeline, average_method):
    
    kfoldcnb = StratifiedKFold(n_splits=splits, shuffle=True, random_state=5)
    accuracycnb = []
    precisioncnb = []
    recallcnb = []
    f1cnb = []
    for train, test in kfoldcnb.split(X, Y):
        cnb_fit = pipeline.fit(X[train], Y[train])
        predictioncnb = cnb_fit.predict(X[test])
        scorescnb = cnb_fit.score(X[test],Y[test])
        class_names = ['Negative','Neutral','Positive']
        cm = confusion_matrix((Y[test]), predictioncnb)
        cnbdf_cm = pd.DataFrame(cm, index=class_names, columns=class_names)
        
        accuracycnb.append(scorescnb * 100)
        print('Accuracy: ', scorescnb * 100)
        precisioncnb.append(precision_score(Y[test], predictioncnb, average=average_method)*100)
        print(' '*10, class_names)
        print('Precision:',precision_score(Y[test], predictioncnb, average=None))
        recallcnb.append(recall_score(Y[test], predictioncnb, average=average_method)*100)
        print('Recall:   ',recall_score(Y[test], predictioncnb, average=None))
        f1cnb.append(f1_score(Y[test], predictioncnb, average=average_method)*100)
        print('F1 score: ',f1_score(Y[test], predictioncnb, average=None))
        print('-'*50)

        cnbaccumean = np.mean(accuracycnb)
        cnbprecmean = np.mean(precisioncnb)
        cnbrecamean = np.mean(recallcnb)
        cnbf1scmean = np.mean(f1cnb)
    print('Accuracy: %.2f%% (+/- %.2f%%)' % (cnbaccumean, np.std(accuracycnb)))
    print('Precision: %.2f%% (+/- %.2f%%)' % (cnbprecmean, np.std(precisioncnb)))
    print('Recall: %.2f%% (+/- %.2f%%)' % (cnbrecamean, np.std(recallcnb)))
    print('F1 score: %.2f%% (+/- %.2f%%)' % (cnbf1scmean, np.std(f1cnb)))
    print(' ')
    print('Confusion matrix\n', cnbdf_cm)
    return accuracycnb, precisioncnb, recallcnb, f1cnb
```

## Buat pipeline untuk mengatasi imbalance dataset dengan metoda Over Sampler

```python
#Complement Naive Bayes Random Over Sampler
#TFIDF
cnb_ROS_tvec1_pipeline = make_pipeline(tvec1, ROSnb, cnb) #unigram
cnb_ROS_tvec2_pipeline = make_pipeline(tvec2, ROSnb, cnb) #bigram
cnb_ROS_tvec3_pipeline = make_pipeline(tvec3, ROSnb, cnb) #trigram
#Count Vectorizer
cnb_ROS_cvec1_pipeline = make_pipeline(cvec1, ROSnb, cnb) #unigram
cnb_ROS_cvec2_pipeline = make_pipeline(cvec2, ROSnb, cnb) #bigram
cnb_ROS_cvec3_pipeline = make_pipeline(cvec3, ROSnb, cnb) #trigram

#Complement Naive Bayes Random Over Sampler
#TFIDF
cnb_SMOTE_tvec1_pipeline = make_pipeline(tvec1, SMOTEnb, cnb) #unigram
cnb_SMOTE_tvec2_pipeline = make_pipeline(tvec2, SMOTEnb, cnb) #bigram
cnb_SMOTE_tvec3_pipeline = make_pipeline(tvec3, SMOTEnb, cnb) #trigram
#Count Vectorizer
cnb_SMOTE_cvec1_pipeline = make_pipeline(cvec1, SMOTEnb, cnb) #unigram
cnb_SMOTE_cvec2_pipeline = make_pipeline(cvec2, SMOTEnb, cnb) #bigram
cnb_SMOTE_cvec3_pipeline = make_pipeline(cvec3, SMOTEnb, cnb) #trigram
```

```python
def get_top_unigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(1, 1)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = get_top_unigram(clean_df['New_Text'], 20)
print('UNIGRAM')
print(' ')
for word, freq in common_words:
    print(word, freq)
df_unigram = pd.DataFrame(common_words, columns = ['New_Text' , 'count'])

def get_top_bigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2, 2)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = get_top_bigram(clean_df['New_Text'], 20)
print('-' * 50)
print('BIGRAM')
print(' ')
for word, freq in common_words:
    print(word, freq)
df_bigram = pd.DataFrame(common_words, columns = ['New_Text' , 'count'])

def get_top_trigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(3, 3)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = get_top_trigram(clean_df['New_Text'], 20)
print('-' * 50)
print('TRIGRAM')
print(' ')
for word, freq in common_words:
    print(word, freq)
df_trigram = pd.DataFrame(common_words, columns = ['New_Text' , 'count'])
```

```python
df_unigram.groupby('New_Text').sum()['count'].sort_values(ascending=False).iplot(
    kind='bar', yTitle='Count', linecolor='black', title='Top Unigram')

df_bigram.groupby('New_Text').sum()['count'].sort_values(ascending=False).iplot(
    kind='bar', yTitle='Count', linecolor='black', title='Top Bigram')

df_trigram.groupby('New_Text').sum()['count'].sort_values(ascending=False).iplot(
    kind='bar', yTitle='Count', linecolor='black', title='Top Trigram')
```
